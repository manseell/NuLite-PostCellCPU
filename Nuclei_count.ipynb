```python
import json
from shapely.geometry import shape

with open("./Output/20-5312 PDL-1 22C3 Ab/cell_detection/cell_detection.geojson", "r") as f:
    data = json.load(f)

totale = 0
for elemento in data:
    geom = shape(elemento["geometry"])
    n = len(geom.geoms)
    classe = elemento["properties"]["classification"]["name"]
    print(f"{classe}: {n} nuclei")
    totale += n

print(f"\nTotale nuclei: {totale}")
```
